```python
import pandas as pd
import matplotlib as plt
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression(random_state=0)
titanic_train=pd.read_csv("train.csv")
validation=titanic_train["Survived"]
titanic_test=pd.read_csv("test.csv")
preproc=titanic_train.drop(["Survived","Name"],axis=1)
preproc1=titanic_test.drop(["Name"],axis=1)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
imputer1=SimpleImputer(missing_values=np.nan,strategy="median")
imputer2=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
X_numtrain=preproc.select_dtypes(include=[np.number])
X_chartrain=preproc.select_dtypes(include=object)
X_numtrain1=X_numtrain.copy()
X_chartrain1=X_chartrain.copy()
imputer1.fit(X_numtrain1)
Y_numtrain=preproc1.select_dtypes(include=[np.number])
Y_chartrain=preproc1.select_dtypes(include=object)
Y_numtrain1=Y_numtrain.copy()
Y_chartrain1=Y_chartrain.copy()
imputer1.fit(X_numtrain1)
```

```python
imputer2.fit(X_chartrain1)
```

```python
imputer2.statistics_
```

```python
X1=imputer1.transform(X_numtrain1)
Y1=imputer1.transform(Y_numtrain1)
```

```python
X_numtrain=pd.DataFrame(X1,columns=X_numtrain.columns,
                         index=X_numtrain.index)
Y_numtrain=pd.DataFrame(Y1,columns=Y_numtrain.columns,
                         index=Y_numtrain.index)
```

```python
X2=imputer2.transform(X_chartrain1)
X_chartrain=pd.DataFrame(X2,columns=X_chartrain.columns,
                         index=X_chartrain.index)
Y2=imputer2.transform(Y_chartrain1)
Y_chartrain=pd.DataFrame(Y2,columns=Y_chartrain.columns,
                         index=Y_chartrain.index)
```

```python
Y_numtrain
```

```python
frames=[X_numtrain,X_chartrain]
X=pd.concat(frames,axis=1,join='outer')
frames1=[Y_numtrain,Y_chartrain]
Y=pd.concat(frames1,axis=1,join='outer')
```

```python
Y
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder=OrdinalEncoder()
X_preenc=X[["Sex","Ticket","Cabin","Embarked"]]
X=X.drop(["Sex","Ticket","Cabin","Embarked"],axis=1)
Y_preenc=Y[["Sex","Ticket","Cabin","Embarked"]]
Y=Y.drop(["Sex","Ticket","Cabin","Embarked"],axis=1)
```

```python
X_enc=ordinal_encoder.fit_transform(X_preenc)
Y_enc=ordinal_encoder.fit_transform(Y_preenc)
```

```python
Y_enc
```

```python
X3=pd.DataFrame(X_enc,columns=X_preenc.columns,index=X_preenc.index)
Y3=pd.DataFrame(Y_enc,columns=Y_preenc.columns,index=Y_preenc.index)
```

```python
X4=X.copy()
Y4=Y.copy()
X4
```

```python
frames=[X4,X3]
X=pd.concat(frames,axis=1,join="outer")
frames1=[Y4,Y3]
Y=pd.concat(frames1,axis=1,join="outer")
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
X_trainnp=std_scaler.fit_transform(X)
Y_trainnp=std_scaler.fit_transform(Y)
```

```python
X_train=pd.DataFrame(X_trainnp,columns=X.columns,index=X.index)
Y_train=pd.DataFrame(Y_trainnp,columns=Y.columns,index=Y.index)
```

```python
X_train
```

```python
clf.fit(X_train,validation)
```

```python
Y_results1=clf.predict(Y)
```

```python
Y_results2=pd.DataFrame(Y_results1,columns=["Survived"])
```
